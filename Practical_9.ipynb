9A:
college professor believes that if the grade for internal examination is high in a class, the grade for external examination will also be high. A random sample of 7 students in that class was selected, and the data is given below. Write a python program for linear regression on the above dataset, and find the coefficients m and b. Predict the external marks if internal marks are 24

#### **Let x = internal marks and y = external marks**

```python
x = [15, 23, 18, 23, 24, 22, 22.0]
x = np.array(x)
y = [49, 63, 58, 60, 58, 61, 60]
n = 7
```

```python
sum_x_y = sum([i*j for i, j in zip(x, y)])
sum_x_square= sum([i**2 for i in x])
sum_y_square = sum([i**2 for i in y])
```

```python
m = ((n*sum_x_y) - (sum(x)*sum(y))) / ( (n*sum_x_square) - sum(x)**2)
```

```python
b = ((sum(y)*sum_x_square) - sum(x)*sum_x_y) / ((n*sum_x_square) - sum(x)**2)
```

```python
plt.scatter(x, y)
y_plt = m*x+b
plt.plot(x, y_plt, color='red')
plt.show()
```

```python
def msc(y_pred, y_actual):
  msc = sum([(i-j)**2 for i, j in zip(y_pred, y_actual)])/len(y_pred)
  return msc
```

```python
y_pred = m*x + b
c = msc(y_pred, y)
print(c)

y_pred = (m-0.1)*x + b
c = msc(y_pred, y)
print(c)

y_pred = (m+0.1)*x + b
c = msc(y_pred, y)
print(c)
```

```python
m1_change = [(m+i) for i in m_change]
```

```python
y_preds = [(m+i)*x+b for i in m_change]
```

```python
msc_arr = [msc(i, y) for i in y_preds]
for i in range(len(msc_arr)):
  print("For m={:.3f}, msc is {:.3f}".format(m1_change[i], msc_arr[i]))
```

```python
plt.plot(m1_change, msc_arr)
plt.xlabel("m")
plt.ylabel("mse")
plt.show()
```

```python
b_change = np.arange(-1.0, 1.1, 0.1)
b1_change = [(b+i) for i in b_change]
yb_preds = [(m)*x+(b+i) for i in b_change]
mscb_arr = [msc(i, y) for i in yb_preds]
plt.plot(b1_change, mscb_arr)
plt.xlabel("b")
plt.ylabel("mse")
plt.show()
```

# 9B
 Generate 1000 points from normal distribution with mean as 100 and standard deviation 20,
store these numbers in a list data1. Generate 1000 points from normal distribution with mean as 50
and standard deviation 10, store these numbers in a list data2. Construct a list data3=data1+data2.
Write a python program for linear regression between dta11 and data3, and find the coefficients m
and b.

```python
seed(2)
data1 = [np.random.randn(1000)*100+20]
```

```python
plt.hist(data1, bins='auto')
plt.show
```

```python
data2 = np.random.randn(1000)*50+10
```

```python
data3 = data1+data2
```

```python
plt.hist(data3, bins='auto')
plt.show
```

```python
def linear_reg(ls1, ls2):
  n = len(ls1)
  X = sum(ls1)
  Y = sum(ls2)
  XY = sum([i*j for i, j in zip(ls1, ls2)])
  X_2 = sum([i**2 for i in ls1])
  m = ((n*XY) - X*Y) / ((n*X_2)-X**2)
  b = ((Y*X_2)-(X * XY)) / ((n*X_2) - X**2)
  return m, b
```

```python
m, b = linear_reg(data1, data3)
```

```python
print("Value of m is: {:.3f}\nValue of b is: {:.3f}".format(m, b))
```

```python
print("Expression is: y = {:.3f}x + {:.3f}".format(m, b))
```

```python
plt.scatter(data1, data3)
x = np.array(data1)
y = np.array(m*data1 + b)
plt.plot(x, y, color='red')
plt.show()
```

```python
def msc(y_pred, y_actual):
  msc = sum([(i-j)**2 for i, j in zip(y_pred, y_actual)])/len(y_pred)
  return msc
```

```python

y_same= m*x + b
c_same = msc(y_same, y)

y_neg= m*x - b
c_neg = msc(y_neg, y)

y_inc= (m+1)*x + b
c_inc = msc(y_inc, y)
```
