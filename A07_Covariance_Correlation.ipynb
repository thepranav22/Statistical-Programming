```python
def emean(arr):
  sum=0
  for i in arr:
    sum+=i
  return (sum/len(arr))
a=[1,2,3]
emean(a)
```

```python
def ecovariance(x,y):
  mean1=emean(x)
  mean2=emean(y)
  sum=0
  if(len(x)==len(y)):
    for i in range(len(x)):
      sum+=((x[i]-mean1)*(y[i]-mean2))
    return (sum/(len(x)-1))
  else:
    return False
a=np.array([1,2,3,6,7,1,8])
b=np.array([4,5,6,2,7,0,1])
if(ecovariance(a,b)):
  print("Covariance:",ecovariance(a,b))
else: 
  print("Length of arrays not equal")
```

```python
np.cov(a,b)
```

```python
def cov_matrix(x,y):
 return np.matrix(([ecovariance(x,x),ecovariance(x,y)],
                [ecovariance(y,x),ecovariance(y,y)]))
```

```python
def stdev(arr):
   mean1=emean(arr)
   sum=0
   for i in arr:
      sum+=((i-mean1)**2)
   return np.sqrt((sum)/(len(arr)))
def corr(x,y):
  return cov_matrix(x,y)/(stdev(x)*stdev(y))
c=corr(a,b)
print(c)
```
