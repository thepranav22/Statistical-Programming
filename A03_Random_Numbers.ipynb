```python
import random

def rsubsample(ele):
  index = random.randint(0, len(ele) - 1)
  subsample = random.sample(ele, index)
  print(subsample)

# List of elements to choose from
elements = ['apple', 'banana', 'cherry', 'date', 'elderberry']

rsubsample(elements)
```

```python
# Write a python program to generate random integers from uniform distributions in
# range of 1 to 6 using randint to simulate dice throwing.


import numpy as np

def simulate_dice_throw():
    # Generate a random integer from a uniform distribution in the range of 1 to 6
    dice_number = np.random.randint(1, 7)
    return dice_number

# Simulate dice throws and print the result
for i in range(10):
    dice_number = simulate_dice_throw()
    print(f"Dice throw {i+1}: {dice_number}")
```
