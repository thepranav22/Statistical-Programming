2A. Write a python program to plot two ideal Gaussian Distributions using norm.pdf with
different mean and standard deviation.

```python
from numpy import arange
from numpy.random import seed
from numpy.random import randn
from matplotlib import pyplot
from numpy import mean
from numpy import var
from numpy import std
from numpy import median
from scipy.stats import norm
```

```python
x_axis = arange(-6,6,0.001)
y_axis = norm.pdf(x_axis,0,1)
y_axis1 = norm.pdf(x_axis,0,2)
pyplot.plot(x_axis,y_axis)
pyplot.plot(x_axis,y_axis1)
pyplot.show()
```

```python
#seed(1)
data=5*randn(10000)+50
pyplot.hist(data,bins=10,color='red')
pyplot.show()
```

```python
#seed(1)
data1=5*randn(1000)+50
#5=S.D.,50=mean
pyplot.hist(data1,bins=10,color='green')
result=median(data1)
print('Median:%.4f'%result)
```

```python
#seed(1)
data=5*randn(1000)+50
data[0]=10000
data[len(data)-1]=10000
pyplot.hist(data,bins=10,color='red')
result=mean(data)
print('Mean:%.4f'%result)
```

```python
#seed(1)
data1=5*randn(1000)+50
pyplot.hist(data1,bins=10,color='red')
result=mean(data1)
print('Mean:%.4f'%result)
```

```python
#seed(1)
data1=5*randn(1000)+50
pyplot.hist(data1,bins=10,color='red')
result=var(data1)
print('Variance:%.4f'%result)
```
