```python
import matplotlib.pyplot as plt
import numpy as np
import numpy.random as rd
import scipy.stats as sp
import math
from sklearn.neighbors import KernelDensity as kd
```

EXPERIMENT 1


```python
def find_mean(arr):
  sum = 0
  for i in arr:
      sum+=i
  return sum/len(arr)
```

```python
def std_dev(x):
  mean_1 = find_mean(x)
  std = 0
  for i in range(len(x)):
    std += (x[i]-mean_1)**2
  return math.sqrt(std/(len(x)-1))
```

```python
rd.seed(1)
sample = rd.normal(50, 5, 10000)
```

```python
plt.hist(sample, bins='auto')
plt.show()
```

```python
samp_mean = find_mean(sample)
samp_std = std_dev(sample)
```

```python
print('mean is: {:.2f}\nstd deviation is: {:.2f}'.format(samp_mean, samp_std))
```

EXPERIMENT 2


```python
dist = sp.norm(samp_mean, samp_std)
```

```python
values = [val for val in range(40, 55)]
```

```python
prob = [dist.pdf(val) for val in values]
```

```python
plt.plot(values, prob)
plt.show()
```

EXPERIMENT 3

```python
sample_1 = rd.normal(20, 5, 300)
sample_2 = rd.normal(40, 5, 700)
```

```python
sample_comb = np.hstack((sample_1, sample_2))
```

```python
plt.hist(sample_comb, bins='auto')
plt.show()
```

```python
model = kd(bandwidth=2, kernel='gaussian')
samp = sample_comb.reshape((len(sample_comb), 1))
```

```python
model.fit(samp)
```

```python
values = np.asarray([value for value in range(1, 60)])
values = values.reshape((len(values), 1))
```

```python
prob = model.score_samples(values)
prob = np.exp(prob)
```

```python
plt.hist(samp, bins=50, density=True)
plt.plot(values[:], prob, color='orange')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate 10000 samples from a normal distribution with mean 0 and standard deviation 1
samples = np.random.normal(0, 1, 10000)

# Plot histograms with different bin values
bins_values = [10, 20, 50, 100]
for bins in bins_values:
    plt.hist(samples, bins=bins, density=True, alpha=0.5, label=f'bins={bins}')

plt.legend()
plt.xlabel('Value')
plt.ylabel('Probability density')
plt.title('Histograms of 10000 samples from a normal distribution')
plt.show()
```
