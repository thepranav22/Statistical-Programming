```python
import numpy as np
import matplotlib.pyplot as plt

# Number of coin tosses
n = 10
# Number of simulations
simulations = 10000

# Counts of heads in each simulation
counts = [0] * 11

for i in range(simulations):
    tosses = np.random.randint(0, 2, n)
    heads = sum(tosses)
    counts[heads] += 1

# Plot the bar graph
plt.bar(range(11), counts)
plt.xlabel("Number of heads")
plt.ylabel("Count")
plt.title("Distribution of heads in 10 coin tosses")
plt.show()
```
